```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.metrics import r2_score
from python_utils import *
```

```python
df = pd.read_csv('dataset/diabetes-dataset.csv', header=(0))
print("Número de linhas e colunas:",df.shape)
df.dropna()
df.head(-10)
```

```python
import seaborn as sns
corr = df.corr()

plt.figure(figsize=(8, 5))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='Reds')
```

1) Relationship between number of Age and Pregnancies.

```python
#Gráfico de dispersão
#Reta com ajustada
data = df.to_numpy()
x = data[:,7] #Age
y = data[:,0] #Pregnancies
X = np.array(x).reshape(-1, 1)
Y = np.array(y)

#Reta com ajustada
modelo = LinearRegression().fit(X, Y)
print(modelo.intercept_)
print (modelo.coef_)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].scatter(X, Y)
axs[0].set_title('Gráfico de Dispersão (Age x Pregnancies)')
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Pregnancies")
axs[0].legend()

axs[1].scatter(X, Y)
axs[1].plot(X, modelo.predict(X), color = 'red')
axs[1].set_title('Gráfico de Ajuste (Age x Pregnancies)')
axs[1].set_xlabel("Age")
axs[1].set_ylabel("Pregnancies")
axs[1].legend()
```

```python
#Coeficiente de correlação de Pearson
correlacao = np.corrcoef(x, y)
print(correlacao)
```

```python
#Coeficiente de Determinação R^2
y_pred=modelo.predict(X)
R2 = r2_score(y, y_pred)
print('R2:', R2)
```
